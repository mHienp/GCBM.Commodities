```python
indonesia = gpd.read_file('/content/IDN_AL2_Indonesia.json')
```

```python
indonesia_planted = gpd.read_file('/content/IDN_Indonesia_SDPT.json')
```

```python
indonesia_planted.head(1)
```

```python
timber = indonesia_planted[indonesia_planted['org_name'].str.contains('timber', case=False, na=False)]
```

```python
timber['org_name'].unique()
```

```python
timber['common_name'] = timber['species_simp']
```

```python
acacia = indonesia_planted[indonesia_planted['common_name'].str.contains('acacia', case=False, na=False)]
```

```python
oilpalm = indonesia_planted[indonesia_planted['common_name'].str.contains('oil palm', case=False, na=False)]
```

```python
rubber = indonesia_planted[indonesia_planted['common_name'].str.contains('rubber', case=False, na=False)]
```

```python
coffee = indonesia_planted[indonesia_planted['common_name'].str.contains('coffee', case=False, na=False)]
```

```python
cacao = indonesia_planted[indonesia_planted['common_name'].str.contains('cacao', case=False, na=False)]
```

```python
commodities_reg = pd.concat([timber, acacia, oilpalm, rubber, coffee, cacao])
```

```python
cmap = 'Spectral_r'
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
ax.set_title('PLANTATIONS IN INDONESIA')

indonesia_planted.plot(ax=ax, column='common_name', cmap=cmap, edgecolor='None', legend=True)
indonesia.plot(ax=ax, color='None', edgecolor='grey', legend=False)

# Move the legend to the right of the map
leg = ax.get_legend()
leg.set_bbox_to_anchor((1.25, 1))  # (horizontal, vertical)

plt.show()
```

```python
cmap = 'Spectral_r'
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
ax.set_title('ACACIA, CACAO, COFFEE, OIL PALM, RUBBER, TIMBER')

commodities_reg.plot(ax=ax, column='common_name', cmap=cmap, edgecolor='None', legend=True)
indonesia.plot(ax=ax, color='None', edgecolor='grey', legend=False)

# Move the legend to the right of the map
leg = ax.get_legend()
leg.set_bbox_to_anchor((1.25, 1))  # (horizontal, vertical)

plt.show()
```
